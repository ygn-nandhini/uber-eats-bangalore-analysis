```python
import pandas as pd
df = pd.read_csv("Uber_Eats_data.csv")
df.head()
```

```python
df['votes'] = pd.to_numeric(df['votes'], errors='coerce')
```

```python
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].astype(str)
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].str.replace(',', '').str.strip()
df['approx_cost(for two people)'] = pd.to_numeric(df['approx_cost(for two people)'], errors='coerce')
```

```python
df['rate'] = df['rate'].astype(str)
df['rate'] = df['rate'].str.replace('/5', '').str.strip()
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')
```

```python
import mysql.connector
import json

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Ygn@1994",
    database="my_project"
)

cursor = conn.cursor()

with open(r'C:\Users\admin\Downloads\orders.json') as f:
    data = json.load(f)

query = """
INSERT INTO orders 
(order_id, restaurant_name, order_date, order_value, discount_used, payment_method)
VALUES (%s, %s, %s, %s, %s, %s)
"""

values = [
    (
        row['order_id'],
        row['restaurant_name'],
        row['order_date'],
        row['order_value'],
        row['discount_used'],
        row['payment_method']
    )
    for row in data
]

cursor.executemany(query, values)
conn.commit()

cursor.close()
conn.close()
```
